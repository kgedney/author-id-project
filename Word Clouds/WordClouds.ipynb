```python
import csv
import numpy as np
from PIL import Image 
from wordcloud import ImageColorGenerator
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
#Function to read txt files
def load(text):
    lines = []
    with open(text) as f:
        reader = csv.reader(f)
        for row in reader:
            lines.append(row)
    return lines
```

```python
#load data
txt_files = load('words_68.txt') #best ID for pure CNN
txt = load('words_1.txt') # best ID for CNN + LSTM
```

```python
#Function to flatten list of lists to single list
flatten = lambda l: [item for sublist in l for item in sublist]
```

```python
#Convert lists to strings
flat = flatten(txt_files)
str1 = ''.join(flat)

flat2 = flatten(txt)
str2 = ''.join(flat2)
```

```python
#Create Wordcloud
from wordcloud import WordCloud
import matplotlib.pyplot as plt

wordcloud = WordCloud(background_color = "white").generate(str1)

plt.figure(figsize=(10,5))
plt.title("Author Best Predicted by CNN")
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
#Create Wordcloud
from wordcloud import WordCloud
import matplotlib.pyplot as plt

wordcloud = WordCloud(background_color = "white").generate(str2)

plt.figure(figsize=(10,5))
plt.title("Author Best Predicted by CNN+LSTM")
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
